```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
y = iris.target
X = iris.data[:, 2:4]
X_length = X[:, 0]
X_width = X[:, 1]
```

## 看一下数据的情况

```python
X
```

```python
y
```

```python
fig1 = plt.figure(1, figsize=(8,8), dpi=100)
plt.scatter(X_length, X_width)
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

## 用聚类试一下

```python
estimator = KMeans(n_clusters=3)
estimator.fit(X)
labels = estimator.labels_

fig2 = plt.figure(2, figsize=(8,8), dpi=100)
plt.scatter(X_length, X_width, c=labels)
plt.xlabel('length')
plt.ylabel('width')
plt.show()
```

## 参照答案，看看机器分得准不准

```python
fig3 = plt.figure(3, figsize=(8,4), dpi=100)

plt.subplot(121)
plt.scatter(X_length, X_width, c=y)
plt.xlabel('length')
plt.ylabel('width')
plt.title('ground truth')

plt.subplot(122)
plt.scatter(X_length, X_width, c=labels)
plt.xlabel('length')
plt.ylabel('width')
plt.title('machine did this')

plt.show()
```
