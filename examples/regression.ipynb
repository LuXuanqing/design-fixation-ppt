```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = [0, 2, 4, 8, 16]
Y = [1, 5, 9, 17, 33]
```

```python
fig1 = plt.figure(1, figsize=(8,5), dpi=100)
plt.scatter(X, Y, color='darkorange')
plt.show()
```

```python
fig2 = plt.figure(2, figsize=(8,5), dpi=100)
plt.scatter(X, Y, color='darkorange')
plt.plot(X, Y, 'c')
plt.show()
```

```python
fig3 = plt.figure(3, figsize=(8,5), dpi=100)
plt.scatter(X, Y, color='darkorange')
plt.plot(X, Y, 'c')
plt.axvline(10, color='y')
plt.show()
```

```python
fig3 = plt.figure(3, figsize=(8,5), dpi=100)
plt.scatter(X, Y, color='darkorange')
plt.plot(X, Y, 'c')
plt.axvline(10, color='y')
plt.scatter(10, 21, color='magenta', linewidth=4)
plt.show()
```
